### 城市發展趨勢

```python
from shapely.geometry import polygon
import geopandas as gpd
```

```python
df2 = df[['x1', 'y1', 'x2', 'y2', "學校加權分數", "醫院加權分數", '交通加權分數', '公園加權分數', '工業區加權分數', '反向效果加權分數', '房源數量_2023', '平均每坪單價_2023','交易總價_2023','score']]
```

```python
taipei = pd.read_csv("year_score/2023_merge_taipei.csv")
newtaipei = pd.read_csv("year_score/2023_merge_new_taipei.csv")

taipei = taipei.replace('#DIV/0!', 0)
newtaipei = newtaipei.replace('#DIV/0!', 0)
```

```python
taipei['score'] = 1*taipei["學校加權分數"] + 1*taipei["醫院加權分數"] + 1*taipei["交通加權分數"]  + 1*taipei["公園加權分數"] + 1*taipei["工業區加權分數"] - 1*taipei["反向效果加權分數"] 
newtaipei['score'] = 1*newtaipei["學校加權分數"] + 1*newtaipei["醫院加權分數"] + 1*newtaipei["交通加權分數"] + 1*newtaipei["公園加權分數"] + 1*newtaipei["工業區加權分數"] - 1*newtaipei["反向效果加權分數"]
```

```python
gdf_point_all = gpd.GeoDataFrame(df2, geometry=gpd.points_from_xy(df2.x1, df2.y1))
gdf_all = gpd.GeoDataFrame(df2, geometry=gpd.points_from_xy(df2.x1, df2.y1))
gdf_all['geometry'] = gdf_all.apply(lambda x: polygon.Polygon([(x['x1'], x['y1']), (x['x2'], x['y1']), (x['x2'], x['y2']), (x['x1'], x['y2'])]), axis=1)
```

```python
gdf_point_t = gpd.GeoDataFrame(taipei, geometry=gpd.points_from_xy(taipei.x1, taipei.y1))
gdf_t = gpd.GeoDataFrame(taipei, geometry=gpd.points_from_xy(taipei.x1, taipei.y1))
gdf_t['geometry'] = gdf_t.apply(lambda x: polygon.Polygon([(x['x1'], x['y1']), (x['x2'], x['y1']), (x['x2'], x['y2']), (x['x1'], x['y2'])]), axis=1)
```

```python
gdf_point_nt = gpd.GeoDataFrame(newtaipei, geometry=gpd.points_from_xy(newtaipei.x1, newtaipei.y1))
gdf_nt = gpd.GeoDataFrame(newtaipei, geometry=gpd.points_from_xy(newtaipei.x1, newtaipei.y1))
gdf_nt['geometry'] = gdf_nt.apply(lambda x: polygon.Polygon([(x['x1'], x['y1']), (x['x2'], x['y1']), (x['x2'], x['y2']), (x['x1'], x['y2'])]), axis=1)
```

### 區域發展分數

```python
# 繪製地圖

fig, ax = plt.subplots(figsize=(6, 6))
gdf_all.plot(ax=ax, column='score', cmap='coolwarm', legend=True, legend_kwds={'label': "score"})
gdf_point_all.boundary.plot(ax=ax)

plt.title('區域發展分數熱度圖')
plt.xlabel('經度')
plt.ylabel('緯度')
plt.show()
```

```python
# 繪製地圖

fig, ax = plt.subplots(figsize=(6, 6))
gdf_nt.plot(ax=ax, column='score', cmap='coolwarm', legend=True, legend_kwds={'label': "score"})
gdf_point_all.boundary.plot(ax=ax)

plt.title('區域發展分數熱度圖-新北市')
plt.xlabel('經度')
plt.ylabel('緯度')
ax.set_xlim([285000, 310000])
ax.set_ylim([2755000, 2780000])
plt.show()
```

```python
# 繪製地圖

fig, ax = plt.subplots(figsize=(6, 6))
gdf_t.plot(ax=ax, column='score', cmap='coolwarm', legend=True, legend_kwds={'label': "score"})
gdf_point_all.boundary.plot(ax=ax)
gdf_point_t.plot(ax=ax, marker='o', color='black', markersize=0.001)

plt.title('區域發展分數熱度圖-台北市')
plt.xlabel('經度')
plt.ylabel('緯度')
plt.show()
```

### 房源數量

```python
# 把房源數量大於 120 的補上平均值
gdf_all['房源數量_2023'] = gdf_all['房源數量_2023'].astype(float)
gdf_all['房源數量_2023'] = gdf_all['房源數量_2023'].apply(lambda x: x if x < 120 else gdf_all['房源數量_2023'].mean())
```
